###  STEP 4. 19년도 1월 19시 ~ 21시 평균 이용객

```python
GNDT_m1
MMDT_m1    
```

```python
GNDT_copy = GNDT_m1.copy()  #1월 데이터 복사
MMDT_copy = MMDT_m1.copy()  
```

```python
GNDT_avg = GNDT_copy.drop(GNDT_copy.columns[range(1,25),], axis=1) # 강남역 1월 날짜와 합계만 추출.
# GNDT_avg
```

```python
MMDT_avg = MMDT_copy.drop(MMDT_copy.columns[range(1,25),], axis=1) # 면목역 1월 날짜와 합계만 추출.
# MMDT_avg
```

```python
# 각역 1월의 요일 도출 
# 0이 월요일, 1이 화, 2이 수, 3가 목, 4가 금, 5이 토, 그리고 6이 일요일
GNDT_week = GNDT_avg['날짜'].dt.dayofweek
GNDT_week
MMDT_week = MMDT_avg['날짜'].dt.dayofweek
MMDT_week
```

```python
# 1월 요일별 강남역 이용자 수 도출.
# 0이 월요일, 1이 화, 2이 수, 3가 목, 4가 금, 5이 토, 그리고 6이 일요일
# i = 0
# a = 0
# b = 0
# for i in list(range(0,7)):
#     a = GNDT_avg[GNDT_week == i]
#     week_GN = a.mean()
    
#     print(a)
#     print(week_GN)
# week_GN.head()
```

```python
GBDT_mon = GNDT_avg[GNDT_week == 0]
GBDT_mon = GBDT_mon.set_index('날짜')
GBDT_mon.rename(columns = {'합 계' : '월요일 평균 :'}, inplace = True)

GBDT_tue = GNDT_avg[GNDT_week == 1]
GBDT_tue = GBDT_tue.set_index('날짜')
GBDT_tue.rename(columns = {'합 계' : '화요일 평균 :'}, inplace = True)


GBDT_wed = GNDT_avg[GNDT_week == 2]
GBDT_wed = GBDT_wed.set_index('날짜')
GBDT_wed.rename(columns = {'합 계' : '수요일 평균 :'}, inplace = True)

GBDT_thu = GNDT_avg[GNDT_week == 3]
GBDT_thu = GBDT_thu.set_index('날짜')
GBDT_thu.rename(columns = {'합 계' : '목요일 평균 :'}, inplace = True)

GBDT_fri = GNDT_avg[GNDT_week == 4]
GBDT_fri = GBDT_fri.set_index('날짜')
GBDT_fri.rename(columns = {'합 계' : '금요일 평균 :'}, inplace = True)

GBDT_sat = GNDT_avg[GNDT_week == 5]
GBDT_sat = GBDT_sat.set_index('날짜')
GBDT_sat.rename(columns = {'합 계' : '토요일 평균 :'}, inplace = True)

GBDT_sun = GNDT_avg[GNDT_week == 6]
GBDT_sun = GBDT_sun.set_index('날짜')
GBDT_sun.rename(columns = {'합 계' : '일요일 평균 :'}, inplace = True)
```

```python
m0 = GBDT_mon.mean()                #각 요일별 평균값 계산
t1 = GBDT_tue.mean()
w2 = GBDT_wed.mean()
t3 = GBDT_thu.mean()
f4 = GBDT_fri.mean()
s5 = GBDT_sat.mean()
s6 = GBDT_sun.mean()
```

```python
GBDT_date_avg = pd.concat([m0,t1,w2,t3,f4,s5,s6])        # 계산한 각 요일의 데이터 병합.
```

```python
print("강남역 요일별 19 ~ 21시 평균 이용객 수")                # 계산값 출력.
GBDT_date_avg
```

```python
GBDT_date_avg = GBDT_date_avg.to_frame()                 # 계산값 프레임화
GBDT_date_avg
```
